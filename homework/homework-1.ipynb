## Task 7: Python Challenge (10 points)

Write a function called `is_even` that:
- Takes a number `n` as input
- Returns `True` if the number is even
- Returns `False` if the number is odd

Then test your function with at least 3 different numbers.

**Hint**: You'll need to learn about:
- How to define a function in Python (`def`)
- The modulo operator (`%`) - what does `n % 2` give you?
- `if`/`else` statements

````python
# Task 7: Write your is_even function here
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

# Test it with at least 3 numbers
print(f'When n is 1, the result is {is_even(1)}')
print(f'When n is 2, the result is {is_even(2)}')
print(f'When n is 3, the result is {is_even(3)}')
````
